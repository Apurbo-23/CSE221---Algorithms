```python
class DSU:
    def __init__(self, n):
        self.parent = list(range(n + 1))
        self.size = [1] * (n + 1)

    def find(self, i):
        if self.parent[i] == i:
            return i
        self.parent[i] = self.find(self.parent[i])
        return self.parent[i]

    def union(self, i, j):
        root_i = self.find(i)
        root_j = self.find(j)

        if root_i != root_j:
            if self.size[root_i] < self.size[root_j]:
                root_i, root_j = root_j, root_i
            self.parent[root_j] = root_i
            self.size[root_i] += self.size[root_j]
            return True
        return False

    def connected(self, i, j):
        return self.find(i) == self.find(j)

n, m = map(int, input().split())
edges = []
for _ in range(m):
    u, v, w = map(int, input().split())
    edges.append((w, u, v))

edges.sort()

# Find the first MST using Kruskal's algorithm
dsu_mst = DSU(n)
mst_cost = 0
mst_edges = []
for weight, u, v in edges:
    if dsu_mst.union(u, v):
        mst_cost += weight
        mst_edges.append((weight, u, v))

# If the graph is not connected, no MST exists
if len(mst_edges) != n - 1:
    print(-1)
else:
    second_mst_cost = float('inf')

    # Construct the adjacency list for the MST to find cycles
    adj = [[] for _ in range(n + 1)]
    for w, u, v in mst_edges:
        adj[u].append((v, w))
        adj[v].append((u, w))

    # Function to find the maximum weight edge on the path between two nodes in the MST
    def find_max_weight_on_path(start, end, parent, max_weight, adj):
        if start == end:
            return 0

        queue = [(start, 0, None)]
        visited = {start}
        path = {}
        path_weight = {start: 0}

        while queue:
            curr, curr_w, prev = queue.pop(0)
            path[curr] = prev

            if curr == end:
                break

            for neighbor, weight in adj[curr]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append((neighbor, weight, curr))
                    path_weight[neighbor] = weight

        max_w = 0
        node = end
        while node != start:
            prev_node = path[node]
            # Find the weight of the edge connecting prev_node and node in the MST
            for neighbor, weight in adj[prev_node]:
                if neighbor == node:
                    max_w = max(max_w, weight)
                    break
            node = prev_node
        return max_w


    # Iterate through edges not in the first MST
    for weight, u, v in edges:
        if (weight, u, v) not in mst_edges:
            # If adding this edge creates a cycle
            if not dsu_mst.connected(u, v): # This check is actually redundant after finding the first MST
                # Find the maximum weight edge on the cycle created by adding (u,v)
                # The cycle is formed by the edge (u,v) and the unique path between u and v in the MST
                max_w_on_path = find_max_weight_on_path(u, v, {}, {}, adj) # Need to pass parent and max_weight correctly
                if max_w_on_path != float('inf'): # If a path exists
                    second_mst_cost = min(second_mst_cost, mst_cost - max_w_on_path + weight)

    # Re-implementing the logic to find the maximum weight edge on the path in the MST
    def find_max_edge_on_path(u, v, adj):
        parent = {}
        max_edge_weight = {}
        queue = [(u, -1, -1)] # (current_node, parent_node, edge_weight_from_parent)
        visited = {u}

        while queue:
            curr, par, weight = queue.pop(0)
            parent[curr] = par
            max_edge_weight[curr] = weight

            if curr == v:
                break

            for neighbor, edge_w in adj[curr]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append((neighbor, curr, edge_w))

        max_w = 0
        node = v
        while node != u:
            prev_node = parent[node]
            # Find the weight of the edge connecting prev_node and node in the MST
            curr_edge_w = -1
            for neighbor, weight in adj[prev_node]:
                if neighbor == node:
                    curr_edge_w = weight
                    break
            max_w = max(max_w, curr_edge_w)
            node = prev_node
        return max_w


    second_mst_cost = float('inf')
    for weight, u, v in edges:
        if (weight, u, v) not in mst_edges:
            max_w_on_path = find_max_edge_on_path(u, v, adj)
            second_mst_cost = min(second_mst_cost, mst_cost - max_w_on_path + weight)


    if second_mst_cost == float('inf'):
        print(-1)
    else:
        print(second_mst_cost)
```
