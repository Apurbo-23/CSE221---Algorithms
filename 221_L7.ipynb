```python
import heapq

def dijkstra(N,graph,S):
    inf = 10**9
    distance = {i: inf for i in range(1,N+1)}
    distance[S] = 0
    stack = [(0,S)]

    while stack:
        current_dist, ver = heapq.heappop(stack)
        if current_dist > distance[ver]:
            continue
        for next,weight in graph[ver]:
            new_dist = current_dist + weight
            if new_dist < distance[next]:
                distance[next] = new_dist
                heapq.heappush(stack,(new_dist,next))
    return distance

N,M,S,T = map(int,input().split())
graph = {i: [] for i in range(1,N+1)}

for _ in range(M):
    u,v,w = map(int,input().split())
    graph[u].append((v,w))

alice_dist = dijkstra(N,graph,S)
bob_dist = dijkstra(N,graph,T)
min_time = float('inf')
meeting_point = -1

for ver in range(1,N+1):
    if alice_dist[ver] == 10**9 or bob_dist[ver] == 10**9:
        continue
    max_time = max(alice_dist[ver],bob_dist[ver])
    if max_time < min_time:
        min_time,meeting_point = max_time,ver
    elif max_time == min_time:
        meeting_point = min(meeting_point,ver)

if min_time == float('inf'):
    print('-1')
else:
    print(min_time,meeting_point)
```

```python
import heapq


N, M = map(int, input().split())

graph = [[] for _ in range(N)]
for _ in range(M):
    u, v, w = map(int, input().split())

    u -= 1
    v -= 1
    graph[u].append((v, w))
    graph[v].append((u, w))


danger_levels = [float('inf')] * N
danger_levels[0] = 0


pq = [(0, 0)]


while pq:
    current_danger, current_city = heapq.heappop(pq)


    if current_danger > danger_levels[current_city]:
        continue


    for neighbor, weight in graph[current_city]:

        new_danger = max(danger_levels[current_city], weight)


        if new_danger < danger_levels[neighbor]:
            danger_levels[neighbor] = new_danger
            heapq.heappush(pq, (new_danger, neighbor))


output = []
for danger in danger_levels:
    if danger == float('inf'):
        output.append("-1")
    else:
        output.append(str(danger))

print(" ".join(output))
```

```python
import heapq

def dijkstra_node_weights():
    N, M, S, D = map(int, input().split())
    node_weights = list(map(int, input().split()))

    S -= 1
    D -= 1

    graph = [[] for _ in range(N)]
    for _ in range(M):
        u, v = map(int, input().split())
        graph[u - 1].append(v - 1)


    distances = [float('inf')] * N
    distances[S] = node_weights[S]

    pq = [(node_weights[S], S)]

    while pq:
        current_cost, current_node = heapq.heappop(pq)

        if current_cost > distances[current_node]:
            continue


        if current_node == D:
            break

        for neighbor in graph[current_node]:

            new_cost = distances[current_node] + node_weights[neighbor]

            if new_cost < distances[neighbor]:
                distances[neighbor] = new_cost
                heapq.heappush(pq, (new_cost, neighbor))

    if distances[D] == float('inf'):
        print(-1)
    else:
        print(distances[D])

dijkstra_node_weights()
```

```python
import heapq

def shortest_distance():
    N, M = map(int, input().split())
    u_edges = list(map(int, input().split()))
    v_edges = list(map(int, input().split()))
    weights = list(map(int, input().split()))


    graph = [[] for _ in range(N)]
    for i in range(M):
        graph[u_edges[i] - 1].append((v_edges[i] - 1, weights[i]))


    distances = [[float('inf')] * 2 for _ in range(N)]

    distances = [[float('inf')] * 2 for _ in range(N)]


    pq = []


    distances = [[float('inf')] * 2 for _ in range(N)]


    distances = [[float('inf')] * 2 for _ in range(N)]


    pq = []


    distances[0][0] = 0
    distances[0][1] = 0
    heapq.heappush(pq, (0, 0, 0))
    heapq.heappush(pq, (0, 0, 1))

    while pq:
        current_dist, current_node, last_edge_parity = heapq.heappop(pq)

        if current_dist > distances[current_node][last_edge_parity]:
            continue


        if current_node == N - 1:
            pass


        for neighbor, weight in graph[current_node]:
            edge_parity = weight % 2


            if last_edge_parity != edge_parity:
                new_dist = current_dist + weight


                if new_dist < distances[neighbor][edge_parity]:
                    distances[neighbor][edge_parity] = new_dist
                    heapq.heappush(pq, (new_dist, neighbor, edge_parity))


    min_dist = min(distances[N - 1][0], distances[N - 1][1])

    if min_dist == float('inf'):
        print(-1)
    else:
        print(min_dist)

shortest_distance()
```

```python
import heapq

def solve():
    N, M, S, D = map(int, input().split())
    S -= 1
    D -= 1

    graph = [[] for _ in range(N)]
    for _ in range(M):
        u, v, w = map(int, input().split())
        u -= 1
        v -= 1
        graph[u].append((v, w))
        graph[v].append((u, w))

    distances = [[float('inf')] * 2 for _ in range(N)]

    pq = [(0, S)]
    distances[S][0] = 0

    while pq:
        d, u = heapq.heappop(pq)

        if d >= distances[u][1]:
            continue

        for v, w in graph[u]:
            new_dist = d + w

            if new_dist < distances[v][0]:
                distances[v][1] = distances[v][0]
                distances[v][0] = new_dist
                heapq.heappush(pq, (distances[v][0], v))
                if distances[v][1] != float('inf'):
                     heapq.heappush(pq, (distances[v][1], v))


            elif new_dist < distances[v][1] and new_dist > distances[v][0]:
                distances[v][1] = new_dist
                heapq.heappush(pq, (distances[v][1], v))


    second_shortest_dist = distances[D][1]

    if second_shortest_dist == float('inf'):
        print(-1)
    else:
        print(second_shortest_dist)

solve()
```
