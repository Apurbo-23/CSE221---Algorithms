```python
from collections import deque

def bfs(n, adj_list):
    visited = [False] * (n + 1)
    queue = deque([1])
    visited[1] = True
    bfs_order = []

    while queue:
        node = queue.popleft()
        bfs_order.append(node)


        neighbors = sorted(adj_list[node])
        for neighbor in neighbors:
            if not visited[neighbor]:
                visited[neighbor] = True
                queue.append(neighbor)

    return bfs_order

n, m = map(int, input().split())
adj_list = [[] for _ in range(n + 1)]

for _ in range(m):
    u, v = map(int, input().split())
    adj_list[u].append(v)
    adj_list[v].append(u)


result = bfs(n, adj_list)
print(*result)
```

```python
from collections import deque

def bfs(start, n, adj_list):
    dist = [-1] * (n + 1)
    parent = [-1] * (n + 1)
    queue = deque([start])
    dist[start] = 0

    while queue:
        node = queue.popleft()

        for neighbor in adj_list[node]:
            if dist[neighbor] == -1:
                dist[neighbor] = dist[node] + 1
                parent[neighbor] = node
                queue.append(neighbor)
    return dist, parent

def solve():
    n, m, s, d, k = map(int, input().split())
    adj_list = [[] for _ in range(n + 1)]

    for _ in range(m):
        u, v = map(int, input().split())
        adj_list[u].append(v)


    dist_s, parent_s = bfs(s, n, adj_list)


    dist_k, parent_k = bfs(k, n, adj_list)

    if dist_s[k] == -1 or dist_k[d] == -1:
        print(-1)
    else:
        print(dist_s[k] + dist_k[d])
        path = []
        curr = k
        while curr != -1:
            path.append(curr)
            curr = parent_s[curr]
        path.reverse()

        curr = d
        temp_path = []
        while curr != -1 and curr != k:
             temp_path.append(curr)
             curr = parent_k[curr]

        temp_path.reverse()
        path.extend(temp_path)

        print(*path)

solve()
```

```python
import sys
from collections import deque

def iterative_dfs_subtree_size(start_node, n, adj_list):

    subtree_size = [0] * (n + 1)

    stack = [(start_node, 0, -1)]


    while stack:
        node, state, parent = stack.pop()

        if state == 0:

            stack.append((node, 1, parent))


            for neighbor in adj_list[node]:
                if neighbor != parent:
                     stack.append((neighbor, 0, node))


        elif state == 1:
            subtree_size[node] = 1


            for neighbor in adj_list[node]:
                 if neighbor != parent:

                     subtree_size[node] += subtree_size[neighbor]


    return subtree_size



n, r = map(int, input().split())


adj_list = [[] for _ in range(n + 1)]
for _ in range(n - 1):
    u, v = map(int, input().split())
    adj_list[u].append(v)
    adj_list[v].append(u)



subtree_size = iterative_dfs_subtree_size(r, n, adj_list)



q = int(input())


for _ in range(q):
    x = int(input())
    print(subtree_size[x])
```

```python
def cycleDetection():
    N,M = map(int,input().split())
    graph_int = [[] for i in range(N+1)]
    degree = [0]*(N+1)
    for j in range(M):
        u,v = map(int,input().split())
        graph_int[u].append(v)
        degree[v] += 1

    que = []
    for k in range(1,N+1):
        if degree[k] == 0:
            que.append(k)
    p=0
    done=0
    while p<len(que):
        temp = que[p]
        p+= 1
        done+= 1
        for nxt in graph_int[temp]:
            degree[nxt] -= 1
            if degree[nxt] == 0:
                que.append(nxt)
    if done>=N:
        print("NO")
    else:
        print("YES")
cycleDetection()
```

```python
def diamond_collection():
    R, H = map(int, input().split())
    grid = [input().strip() for i in range(R)]
    visited = [[False] * H for j in range(R)]
    moves = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    diamond = 0

    for i in range(R):
      for j in range(H):
        if visited[i][j]== False and grid[i][j]!= "#":
          count,stack= 0,[(i,j)]
          visited[i][j]= True
          while stack:
            gr,gc = stack.pop()
            count += grid[gr][gc] == "D"
            for mr,mc in moves:
              nr,nc = gr+mr,gc+mc
              if 0<=nr<R and 0<=nc<H and visited[nr][nc]== False and grid[nr][nc]!= "#":
                visited[nr][nc] = True
                stack.append((nr,nc))
          diamond = max(diamond, count)

    print(diamond)

diamond_collection()
```
